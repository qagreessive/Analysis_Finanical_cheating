```python
import pandas as pd
import numpy as np
```

```python
# 读取原始数据表
pure_dt2=pd.read_csv('pure_dt2.csv')
```

```python
pure_dt2.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
pure_dt2
```

```python
pd1=pd.read_excel('附件1.xlsx')
pd1
```

```python
pd1.nunique()
```

```python
pd1_groups=pd1.groupby(pd1['所属行业'])
pd1_groups.count().sort_values(by='股票代码',ascending=False)
```

```python
pd_tec=pd1.loc[pd1['所属行业']=='信息传输、软件和信息技术服务业']
pd_tec=pd_tec.rename(columns={'股票代码':'TICKER_SYMBOL','所属行业':'INFO'})
pd_tec
```

```python
dt_info=pd.merge(pure_dt2,pd_tec,on='TICKER_SYMBOL')
dt_info.drop(['INFO','TICKER_SYMBOL'],axis=1,inplace=True)
dt_info
```

```python
# 对原表格进行拆分，单独拆分出FLAG，为后续对特征项标准化作准备
dt_info_Flag=dt_info.iloc[:,dt_info.columns=='FLAG']
dt_info_Flag
```

```python
dt_info.drop('FLAG',axis=1,inplace=True)
dt_info
```

```python
dt_info_columns=dt_info.columns
dt_info_columns
```

```python
dt_info.fillna(0,inplace=True)
dt_info
```

```python
# 对特征向量进行标准化
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
dt_info_std=std.fit_transform(dt_info)
dt_info_std=pd.DataFrame(dt_info_std,columns=dt_info_columns)
dt_info_std
```

```python
dt_info_std['FLAG']=dt_info_Flag
```

```python
dt_info_std
```

```python
dt_predict=dt_info_std.loc[dt_info_std['FLAG'].isna()]
dt_predict
```

```python
dt_train_and_test=dt_info_std.loc[dt_info_std['FLAG'].notna()]
dt_train_and_test
```

```python
dt_predict.to_csv('predict_信息传输软件和信息技术服务业.csv')
dt_train_and_test.to_csv('train_and_test_信息传输软件和信息技术服务业.csv')
```
