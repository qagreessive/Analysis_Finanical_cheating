```python
import pandas as pd
import numpy as np
```

```python
X=pd.read_csv('制造业标准化后的特征向量集.csv')
y=pd.read_csv('FLAG标签列.csv')
```

```python
X
```

```python
X.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
X
```

```python
y
```

```python
y.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
y
```

```python
y.value_counts()
```

```python
# y列数据不平衡，不利于学习，通过SMOTE过采样算法解决数据不平衡问题
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
```

```python
y_res.value_counts()
```

```python
X_res
```

```python
y_res
```

```python
# 按7：3比例随机划分试练集与测试集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=0)
```

```python
# 采用集成学习梯度提升决策树进行学习
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
gbm0 = GradientBoostingClassifier(random_state=10)
gbm0.fit(X_train, y_train)
y_pred = gbm0.predict(X_test)
y_predprob = gbm0.predict_proba(X_test)[:,1]
print("Accuracy : %.4g" % metrics.accuracy_score(y_test.values, y_pred))
print("Recall: %.4g"%metrics.recall_score(y_test.values, y_pred))
print("AUC Score (Train): %f" % metrics.roc_auc_score(y_test, y_predprob))
```

```python
# 对梯度提升决策树学习的重要标签进行显示
from collections import OrderedDict
import matplotlib.pyplot as plt
d = {}
for i in range(len(gbm0.feature_importances_)):
    if gbm0.feature_importances_[i] > 0.02:
        d[i] = gbm0.feature_importances_[i]

sorted_feature_importances = OrderedDict(sorted(d.items(), key=lambda x:x[1], reverse=True))
D = sorted_feature_importances
rects = plt.bar(range(len(D)), D.values(), align='center')
plt.xticks(range(len(D)), D.keys(),rotation=90)
plt.show()
```

```python
# 使用决策树模型进行训练
from sklearn.tree import DecisionTreeClassifier #决策树
clf = DecisionTreeClassifier(random_state=0)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
y_predprob = clf.predict_proba(X_test)[:,1]
print("Accuracy : %.4g" % metrics.accuracy_score(y_test.values, y_pred))
print("Recall: %.4g"%metrics.recall_score(y_test.values, y_pred))
print("AUC Score (Train): %f" % metrics.roc_auc_score(y_test, y_predprob))
```

```python
# 显示决策树模型训练出的重要标签
from collections import OrderedDict
import matplotlib.pyplot as plt
d = {}
for i in range(len(clf.feature_importances_)):
    if clf.feature_importances_[i] > 0.02:
        d[i] = clf.feature_importances_[i]

sorted_feature_importances = OrderedDict(sorted(d.items(), key=lambda x:x[1], reverse=True))
D_clf = sorted_feature_importances
rects = plt.bar(range(len(D_clf)), D_clf.values(), align='center')
plt.xticks(range(len(D_clf)), D_clf.keys(),rotation=90)
plt.show()
```

```python
# 使用随机森林进行训练学习
from sklearn.ensemble import RandomForestClassifier #集成学习中的随机森林
rfc = RandomForestClassifier(random_state=0)
rfc = rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
y_predprob = rfc.predict_proba(X_test)[:,1]
print("Accuracy : %.4g" % metrics.accuracy_score(y_test.values, y_pred))
print("Recall: %.4g"%metrics.recall_score(y_test.values, y_pred))
print("AUC Score (Train): %f" % metrics.roc_auc_score(y_test, y_predprob))
```

```python
# 显示随机森林模型训练出的重要标签
from collections import OrderedDict
import matplotlib.pyplot as plt
d = {}
for i in range(len(rfc.feature_importances_)):
    if rfc.feature_importances_[i] > 0.02:
        d[i] = rfc.feature_importances_[i]

sorted_feature_importances = OrderedDict(sorted(d.items(), key=lambda x:x[1], reverse=True))
D_rfc = sorted_feature_importances
rects = plt.bar(range(len(D_rfc)), D_rfc.values(), align='center')
plt.xticks(range(len(D_rfc)), D_rfc.keys(),rotation=90)
plt.show()
```

```python
# 为了保证模型的准确性，采用三个训练模型的交集，影响财务造假的指标为Features_Intersection
D_clf_set=set(D_clf.keys())
D_rfc_set=set(D_rfc.keys())
D_gbm0_set=set(D.keys())
Intersection=D_clf_set.intersection(D_clf_set,D_rfc_set,D_gbm0_set)
Features_Intersection=X_res.columns.values[list(Intersection)]
Features_Intersection
```
