```python
import pandas as pd
import numpy as np
```

```python
#读取原数据表
dt2=pd.read_csv('附件2.csv')
```

```python
dt2.info()
```

```python
#显示表的基本信息
dt2
```

```python
# 因原表特征较多，干扰项较大，先采用挑出只有单个值的特征项来删除
u1 = dt2.nunique()
u1=u1[u1.values<10]
u1=pd.DataFrame(u1.values,u1.index,columns=['Unique'])
u1
```

```python
# c1为dt2表每个column列的有效值的个数
c1=dt2.count()
c1=pd.DataFrame(c1.values,index=c1.index,columns=['Count'])
c1
```

```python
# 取二者交集
u1=u1.join(c1)
u1
```

```python
# 取出仅有一个值的column
str_unique_1=u1.loc[u1['Unique']==1].index
str_unique_1
```

```python
# 删除仅有一个值的column，该column仅含一个值，作为特征项无意义
dt2.drop(str_unique_1,axis=1,inplace=True)
dt2
```

```python
# 取出缺失率大于0.7的column
str_missing_rate=((dt2.isnull().sum())/dt2.shape[0])
str_missing_rate=str_missing_rate[str_missing_rate.values>0.7]
str_missing_rate
```

```python
# 删除缺失率大于0.7的column，因为缺失值太多，作为特征项无意义
dt2.drop(str_missing_rate.index,axis=1,inplace=True)
```

```python
# 看看是否具有object类型的column,在学习中要删除
dt2.info()
```

```python
# 删除dt2表格中Flag列中为NAN的所有行
dt2=dt2.drop(dt2.loc[dt2['FLAG'].isnull()].index,axis=0)
dt2
```

```python
dt2.isnull().sum()
```

```python
# 用0值填充缺失值
dt2.fillna(0,inplace=True)
dt2.isnull().sum()
```

```python
# 读取另一个表，筛选行业
dt1=pd.read_excel("附件1.xlsx")
dt1
```

```python
dt1=dt1.loc[dt1['所属行业']=='制造业']
```

```python
dt1
```

```python
dt1=dt1.rename(columns={'股票代码':'TICKER_SYMBOL','所属行业':'TMT'})
dt1
```

```python
dt1.info()
```

```python
# 将两表根据TICKER_SYMBOL作为关联值合并，便于后续分析
tmt_all_dt=pd.merge(dt2,dt1,on='TICKER_SYMBOL')
tmt_all_dt
```

```python
# 合并后的表都是制造业的，股票代码与所属行业列已经没有意义，因此删除
tmt_all_dt.drop(['TICKER_SYMBOL','TMT'],axis=1,inplace=True)
tmt_all_dt
```

```python
# 选取除FLAG以外的所有列作为特征
tmt_X_dt = tmt_all_dt.iloc[:, tmt_all_dt.columns != 'FLAG']
tmt_X_dt
```

```python
# 取FLAG列作为标签
tmt_y_dt = tmt_all_dt.iloc[:, tmt_all_dt.columns == 'FLAG']
tmt_y_dt
```

```python
# 取X向量集的Columns列名，因为标准化以后就没有列名了
tmt_X_dt_columns=tmt_X_dt.columns
tmt_X_dt_columns
```

```python
# 对特征向量进行标准化
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
tmt_X_dt_std = std.fit_transform(tmt_X_dt)
tmt_X_dt_std = pd.DataFrame(tmt_X_dt_std)
tmt_X_dt_std
```

```python
tmt_X_dt_std.columns=tmt_X_dt_columns
```

```python
tmt_X_dt_std
```

```python
# 查看y值的情况
tmt_y_dt.value_counts()
```

```python
# 将数据存入表格备用
tmt_X_dt_std.to_csv('制造业标准化后的特征向量集.csv')
tmt_y_dt.to_csv('FLAG标签列.csv')
```
